# Basic elements II

We now want to use the script plot_distributions.py in this notebook.

We can simply import it as any other package

```python
import matplotlib.pylab as plt
import plot_distributions
```

```python
# Magically, it imports all the objects of the script as its own METHODS
print(plot_distributions.mainDir)
print(plot_distributions.distributions)
print(plot_distributions.plot_dis) # The function, too!
```

```python
# Now we can call the function easily
plot_distributions.plot_dis("Energy")
```

```python
# Let's plot the 3 distributions all together
fig, axs = plt.subplots(1, 3, figsize=(12,4))
plot_distributions.plot_dis("Energy", ax=axs[0])
plot_distributions.plot_dis("Duration", ax=axs[1])
plot_distributions.plot_dis("Size", ax=axs[2])
```

Not very pythonic, by the way.

We can make the property of the function to plot the data into an axis we pass to.

```python
# Let's plot the 3 distributions all together
fig, axs = plt.subplots(1, 3, figsize=(15,3))
my_choices = ['Energy', 'Size', 'Duration'] # Note the use of plural (my_choices)
for i, my_choice in enumerate(my_choices): # Note the use of singular (my_choice)
    # We need to explicity write ax=... Why?
    plot_distributions.plot_dis(my_choice, ax=axs[i]) 
```
